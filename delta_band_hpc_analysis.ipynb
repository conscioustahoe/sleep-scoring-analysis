# Delta Band HPC Analysis

**Delta Band**: In neuroscience and electroencephalography (EEG), brain activity is categorized into different frequency bands based on the oscillation patterns of electrical activity in the brain. The delta band typically ranges from 0.5 to 4 Hertz (Hz) or cycles per second. Delta activity is characterized by slow, high-amplitude oscillations and is commonly associated with deep sleep and unconsciousness.

**HPC (Hippocampus)**: The hippocampus is a critical region located deep within the brain's temporal lobe. It plays a crucial role in various cognitive functions, including memory formation and spatial navigation. Researchers often study the neural activity in the hippocampus to better understand memory processes and other related functions.

## Load DataFrame

```python
import numpy as np
import pandas as pd
```

```python
data = delta_band_hpc['DeltaBand']['data']
timesteps = delta_band_hpc['DeltaBand']['timestamps']
```

```python
flat_data = np.concatenate(data.all())
df = pd.DataFrame({'DeltaBandHPC': flat_data})
```

```python
df['Timestep'] = range(1, len(timesteps.all()) + 1)
df.set_index('Timestep', inplace=True)
```

## Plot Timeseries

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['DeltaBandHPC'], label='DeltaBandHPC', color='blue')
plt.title('Time Series Plot of DeltaBandHPC')
plt.xlabel('Timestep')
plt.ylabel('DeltaBandHPC Value')
plt.legend()
plt.grid(True)
plt.show()
```
