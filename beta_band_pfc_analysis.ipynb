# Beta Band PFC Analysis

**Beta Band**: In neuroscience and electroencephalography (EEG), brain activity is often categorized into different frequency bands based on the oscillation patterns of electrical activity in the brain. Beta is one of these frequency bands and typically ranges from 13 to 30 Hertz (Hz) or cycles per second. Beta activity is associated with various cognitive processes, including motor control, decision-making, and active thinking.

**PFC (Prefrontal Cortex)**: The prefrontal cortex is a region of the brain located at the front of the frontal lobe. It is involved in higher-order cognitive functions such as executive function, decision-making, working memory, and goal setting. The PFC plays a crucial role in regulating and orchestrating various cognitive processes.

## Load DataFrame

```python
import numpy as np
import pandas as pd
```

```python
data = beta_band_pfc['BetaBand']['data']
timesteps = beta_band_pfc['BetaBand']['timestamps']
```

```python
flat_data = np.concatenate(data.all())
df = pd.DataFrame({'BetaBandPFC': flat_data})
```

```python
df['Timestep'] = range(1, len(beta_band_pfc['BetaBand']['timestamps'].all()) + 1)
df.set_index('Timestep', inplace=True)
```

## Plot Timeseries

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['BetaBandPFC'], label='BetaBandPFC', color='blue')
plt.title('Time Series Plot of BetaBandPFC')
plt.xlabel('Timestep')
plt.ylabel('BetaBandPFC Value')
plt.legend()
plt.grid(True)
plt.show()
```
