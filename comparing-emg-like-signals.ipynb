```python
from scipy.io import loadmat
lfp_buz_features_13_19_27 = loadmat('synthesised-features/lfp_buz_features.mat')
lfp_buz_features_rgs14 = loadmat('synthesised-features/lfp_buz_features_rgs14.mat')
```

```python
import pandas as pd
import numpy as np

lfp_matrix_13_19_27 = np.array(lfp_buz_features_13_19_27['lfpFeatures'])
lfp_matrix_rgs14 = np.array(lfp_buz_features_rgs14['lfpFeatures'])

emg_like_df_13_19_27 = pd.DataFrame(lfp_matrix_13_19_27[:, 4], columns=["EMGLike"])
emg_like_df_rgs14 = pd.DataFrame(lfp_matrix_rgs14[:, 4], columns=["EMGLike"])

timesteps_13_19_27 = 10820
timesteps_rgs14 = 4216
```

```python
emg_like_df_13_19_27['Timestep'] = range(1, timesteps_13_19_27 + 1)
emg_like_df_13_19_27.set_index('Timestep', inplace=True)

emg_like_df_rgs14['Timestep'] = range(1, timesteps_rgs14 + 1)
emg_like_df_rgs14.set_index('Timestep', inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(emg_like_df_13_19_27.index, emg_like_df_13_19_27['EMGLike'], label='EMGLike 13_19_27', color='blue')
plt.title('Time Series Plot of EMGLike 13_19_27')
plt.xlabel('Timestep')
plt.ylabel('EMGLike Value')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(emg_like_df_rgs14.index, emg_like_df_rgs14['EMGLike'], label='EMGLike RGS14', color='green')
plt.title('Time Series Plot of EMGLike RGS14')
plt.xlabel('Timestep')
plt.ylabel('EMGLike Value')
plt.legend()
plt.grid(True)
plt.show()
```
